```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/20111123/13301104001.csv', error_bad_lines=False)
df.head()
```

```python
# drop irrelevant cols
to_drop = ['4', '50#']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
# AYJ seems redundent
print (sum (df[:]['AYJ'] != 'AYJ'))
# remove it
df.drop(['AYJ'], inplace=True, axis=1)
```

```python
df.head()
```

```python
# renaming cols
rename_mapping_cands = ['point_id', 'taxi_id', 'time', 'y', 'x', 'OSM_st', 'OSM_ed', 'speed', 'theta', 'status']
rename_mapping = {}
for i, name in enumerate(df.columns.values):
    rename_mapping[name] = rename_mapping_cands[i]
```

```python
rename_mapping
```

```python
df.rename(rename_mapping, axis=1, inplace=True)
```
