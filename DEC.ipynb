```python
import os
import sklearn
from sklearn import decomposition
import pandas as pd
import numpy as np
from numpy import matmul
import matplotlib
from matplotlib import pyplot as plt
from DEC import for_all_files


%matplotlib inline
```

```python
baseURL = './data2/'
fileurls = os.listdir(baseURL)
for i in range(len(fileurls)):
    fileurls[i] = baseURL + fileurls[i]
```

```python
fileurls
```

```python
df = for_all_files(fileurls)
```

```python
df.shape
```

```python
n_features = 3
max_iter = 1000
nmf = decomposition.NMF(n_components=n_features, max_iter=max_iter)
```

```python
W = nmf.fit_transform(df)
H = nmf.components_
loss = nmf.reconstruction_err_
```

```python
times = [i for i in range(1, 97)]
plt.xlabel("Time interval")
plt.ylabel("Traffic flow")
plt.plot(times, H[:][0], color='r')
plt.plot(times, H[:][1], color='g')
plt.plot(times, H[:][2], color='b')
plt.show()
```
