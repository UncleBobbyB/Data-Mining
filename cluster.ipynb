```python
import sklearn
import pandas as pd
import numpy as np
from numpy import matmul
import matplotlib
from matplotlib import pyplot as plt
from DEC import for_all_files
from sklearn import decomposition, cluster
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
df = for_all_files(['./data2/20111117.csv'])
```

```python
df.head()
```

```python
df.shape
```

```python
n_features = 3
max_iter = 1000
nmf = decomposition.NMF(n_components=n_features, max_iter=max_iter)
```

```python
W = nmf.fit_transform(df)
H = nmf.components_
```

```python
# compute k means
clf = cluster.KMeans(n_clusters=100)
result = clf.fit_predict(W)
```

```python
ax = Axes3D(plt.figure())
ax.scatter(W[:, 0], W[:, 1], W[:, 2], c=result)
```
